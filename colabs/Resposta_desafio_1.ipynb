```python
import sys
from collections import deque

def bfs(grid, N, M, sx, sy, dx, dy):
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # Up, Down, Left, Right
    queue = deque([(sx, sy, 0, [(sx, sy)])])  # (x, y, distance, path)
    visited = set()
    visited.add((sx, sy))

    while queue:
        x, y, dist, path = queue.popleft()

        if (x, y) == (dx, dy):
            print("Steps to reach the destination:")
            for step in path:
                print(step)
            return dist

        for d_x, d_y in directions:
            nx, ny = x + d_x, y + d_y
            if 0 <= nx < N and 0 <= ny < M and grid[nx][ny] == '.' and (nx, ny) not in visited:
                visited.add((nx, ny))
                queue.append((nx, ny, dist + 1, path + [(nx, ny)]))

    print("No path found")
    return -1  # No path found
```

```python
def main():
    # Leitura da entrada
    N, M = map(int, input().split())
    grid = [input().strip().split() for _ in range(N)]
    sx, sy, dx, dy = map(int, input().split())

    # Solução do problema
    result = bfs(grid, N, M, sx, sy, dx, dy)
    print(result)

if __name__ == "__main__":
    main()
```
