```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

w = np.linspace(-4, 4, 100)

# função custo J(w) = 70 - 15w^2 + 0,5w^3 + w^4
def J(w):
    return 70.0 - 15.0*w**2 + 0.5*w**3 + w**4

# derivada agébrica da função custo dj(w)/dw = -30w + 1,5w^2 + 4w^3
def dJdw(w):
    return - 30.0*w + 1.5*w**2 + 4*w**3
```

```python
plt.figure(figsize=(7, 4))
plt.plot(w, J(w))
plt.title("função custo J(w)")
plt.xlabel('w')
plt.ylabel('J(w)')
plt.show()
```

```python
plt.figure(figsize=(7, 4))
plt.plot(w, dJdw(w))
plt.axhline(0, color='black')
plt.title("Derivada algébrica da função custo dJ(w)/dw")
plt.xlabel('w')
plt.ylabel('dJ(w)/dw')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.plot(w, J(w), label='J(w)')
plt.plot(w, dJdw(w), label='dJ(w)/dw')
#plt.axhline(0, color='black', linewidth=1)
plt.title("Função Custo J(w) e a derivada dJ(w)/dw")
plt.xlabel('w')
#plt.ylabel('dJ(w)/dw')
plt.legend()
plt.grid()
plt.show()
```

```python
# derivada numérica
aa=[]
for i in range(len(w)):
    aa.append(J(w[i]))

a=np.array(aa)
dfdw = np.diff(a)/np.diff(w)

plt.figure(figsize=(7, 5))
plt.plot(w[1:], dfdw)
plt.title("dJ(w)/dw - derivada numérica")
plt.xlabel('w')
plt.ylabel('dJ(w)/dw')
plt.grid()
plt.show()
```

```python
# Global minimum - Gradient Descent demonstration
w_0 = -0.1
lr = 0.01
iterations = 20
w_ = w_0
ws = [w_]

for i in range(iterations):
    step = lr * dJdw(w_)
    w_ -= step
    ws.append(w_)

ws = np.array(ws)
plt.figure(figsize=(10, 6))
plt.plot(w, J(w))
plt.plot(ws, J(ws), 'o')
plt.plot(w_0, J(w_0), 'or')
plt.legend(["J(w)", "steps", "starting point"])
plt.title('Global minimum - Gradient Descent demonstration')
plt.xlim(-4.2, 3.5)
plt.ylim(-1, 80)
plt.show()
```

```python
# Local minimum - Gradient Descent demonstration
w_0 = 3.9
lr = 0.001
iterations = 40
w_ = w_0
ws = [w_]

for i in range(iterations):
    step = lr * dJdw(w_)
    w_ -= step
    ws.append(w_)

ws = np.array(ws)
plt.figure(figsize=(8, 5))
plt.plot(w, J(w))
plt.plot(ws, J(ws), 'o')
plt.plot(w_0, J(w_0), 'or')
plt.legend(["J(w)", "steps", "starting point"])
plt.title('Local minimum - Gradient Descent demonstration')
#plt.xlim(-4.2, 1.5)
plt.show()
```
